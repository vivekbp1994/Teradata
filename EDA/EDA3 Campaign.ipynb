```python
df=pd.read_csv('Campaign.csv',encoding="iso-8859-1")
df.head()
```

```python
df.isnull().sum()
```

```python
null_cnt=pd.DataFrame(df.isnull().sum())
null_cnt.reset_index(inplace=True)
null_cnt.columns=['Columns','NULL']
#null_cnt[null_cnt['NULL']>60000]['Columns'].tolist()
#null_cnt.to_csv('Campaign_NULL.csv')
```

```python
df_filt=df[null_cnt[null_cnt['NULL']<=df.shape[0]/2]['Columns'].tolist()]
df_filt.head()
```

```python
df_filt.isnull().sum().sort_values(ascending=False)
```
